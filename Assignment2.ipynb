```python
# Task 1
try:
    order_amount=int(input("Enter the order amount"))
    tax_rate=5
    if order_amount >=2000:
        discount= 15
        subtotal= order_amount-((order_amount* discount )/100)
        tax = subtotal*(tax_rate/100)
        print("subtotal price :",subtotal)
        print("Tax :",tax)
        print("Total Bill :",subtotal+tax)
    
    elif order_amount >=1500 and order_amount<2000:
        discount =10
        subtotal= order_amount-((order_amount*discount)/100)
        tax = subtotal*(tax_rate/100)
        print("subtotal price :",subtotal)
        print("Tax :",tax)
        print("Total Bill :",subtotal+tax)

    elif order_amount >= 1000 and order_amount<1500:
        discount=7
        subtotal= order_amount-((order_amount*discount )/100)
        tax = subtotal*(tax_rate/100)
        print("subtotal price :",subtotal)
        print("Tax :",tax)
        print("Total Bill :",subtotal+tax)

    else:
        discount=0
        print("No Discount, Final price :",order_amount)

except ValueError:
    print("Invalid input")
```

# Task 2

```python
total_revenue=[]
count=0
orders=[1200,2500,800,1750,3000]
for order_amount in orders:
    if order_amount >=2000:
        discount=15
        count+=1
    elif order_amount >=1500 and order_amount<2000:
        discount=10
        count+=1        
    elif order_amount>=1000 and order_amount<1500:
        discount=7
        count+=1
    else:
        discount=0
    final_amount= order_amount-((order_amount*discount)/100)
    total_revenue.append(final_amount)
    
    print(f"{order_amount} --> {discount} --> {final_amount}")

print("Total Revenue after discounts:",sum(total_revenue))
print("No. of orders that received discount:",count)
```

# Task 3

```python

total_revenue=[]
count=0
orders=[1200,2500,800,1750,3000]
while(True):
    menu=input("Enter your choice 1 2 or q")
    if menu =='1':
        a=int(input("Enter the amount-order"))
        orders.append(a)
        continue
    elif menu=='2':
            
        for order_amount in orders:
            if order_amount >=2000:
                discount=15
                count+=1
            elif order_amount >=1500 and order_amount<2000:
                discount=10
                count+=1        
            elif order_amount>=1000 and order_amount<1500:
                discount=7
                count+=1
            else:
                discount=0
            final_amount= order_amount-((order_amount*discount)/100)
            total_revenue.append(final_amount)
            
            print(f"{order_amount} --> {discount} --> {final_amount}")
        continue

    elif menu=='q':
        print("program exit")
        break
    else:
        print("Invalid input")
        continue

```
