## Task1

```python
def apply_discount(price,discount_percent=5):

    if discount_percent<=60:
        amount=price-((price*discount_percent)/100)
        return amount
    else:
        return "discount percent exceeds 60 % "

print(apply_discount(1000,10))
print(apply_discount(500))
print(apply_discount(1000,60))
print(apply_discount(1000,59))
print(apply_discount(1000,70))
```

## Task 2

```python
def factorial(n):
    if n>=0:
        if n==0 or n==1:
            return 1
        result=n*factorial(n-1)
        return result
    else:
        return "number is negative"

print(factorial(5))
print(factorial(0))
print(factorial(-3))
        
```

## Task 3

```python
# to calculate the gst
gst=lambda price : price+(price*0.18)

# to calculate the gst after discount

discount = lambda price, discount : price-(price*(discount/100)) + (price-(price*(discount/100)))*0.18

print(gst(100))
print(discount(100,10))
```

## Task 7

```python
prices_list=[]
def add_prices(prices_list,prices):
    prices=int(input("Enter the price"))
    prices_list.append(prices)
    print(prices_list)

def get_average_price(prices_list):
    print(sum(prices_list)/len(prices_list))

def get_max_price(prices_list):
    print(max(prices_list))

while True:
    print("""\t1 --> Add price\n\t2 --> Show average price\n\t3 --> maximum price\n\tq --> quit""")
    choice=input("Enter your choice")

    if choice =='1':
        add_prices(prices_list,prices)
    elif choice =='2':
        get_average_price(prices_list)
    elif choice =='3':
        get_max_price(prices_list)
    elif choice =='q':
        print("Program Terminated")
        break
    else:
        print("Invalid input")
        continue
```
