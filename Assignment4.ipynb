## Task 5

```python
products = []

for i in range(3):
    name = input(f"Enter product {i+1} name: ")
    price = float(input(f"Enter price of {name}: "))
    products.append((name, price))


with open("products.txt", "w") as f:
    for name, price in products:
        f.write(f"{name} | {price}\n")


with open("products.txt", "r") as f:
    for line in f:
        name, price = line.strip().split("|")
        print(f"Product: {name.strip()}  -->  Price:{price.strip()}")
```

## Task 7

```python
prices={"Mouse":500,"Keyboard":8000,"Monitor":7000,"Pendrive":400,"Camera":5000}
discount_percent=int(input("Enter the discount price"))


with open("discount_report.txt", "w") as f:
    
    total_discounted = 0
    
    for product, price in prices.items():
        discounted_price = price - (price * discount_percent / 100)
        total_discounted += discounted_price
        
        f.write(f"{product} | {price} | {discounted_price:.2f}\n")
    
    # Step 4: Write Summary
    total_items = len(prices)
    average_discounted = total_discounted / total_items
    
    f.write("\n")
    f.write("Summary\n")
    f.write(f"Total Items: {total_items}\n")
    f.write(f"Average Discounted Price: {average_discounted:.2f}\n")


with open("discount_report.txt", "r") as f:
    print(f.read())
```
