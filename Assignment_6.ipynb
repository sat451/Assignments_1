## Task 1

```python
try:
    num=int(input("Enter the numerator"))
    den=int(input("Enter the denominator"))
    result=num/den
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Invalid Input")
else:
    print("Result:",result)
finally:
    print("Operation completed")
```

## Task 2

```python
prices=[120,350,'abc',500,-200,800]
sum=0
for i in prices:
    try:
        if i>0:
            sum+=i
        else:
            raise ValueError
        
    except TypeError:
        pass
        
    else:
        print(sum)
    
```

```python
prices = [120, 350, 'abc', 500, -200, 800]
total = 0

for i in prices:
    try:
        if i > 0:
            total += i
        else:
            raise ValueError("Negative price not allowed")
    
    except TypeError:
        print("Not a number")
    
    except ValueError:
        print("Skipping")
    
    else:
        print("Running Total:", total)

print("Final Total:", total)
```

## Task 3

```python
def check_age(age):
    if age<1 or age>120:
        raise ValueError("Age must be between 1 and 120")
    return "valid age"

try:
    age=int(input("Enter the age"))

    print(check_age(age))

except ValueError as ve:
    print(ve)
    
```

```python
try:
    filename = input("Enter the filename: ")


    with open(filename, "r") as f:
        print("\nFirst 3 lines of the file:\n")
        
        for i in range(3):
            line = f.readline()
            if not line:   # Stops if file has fewer than 3 lines
                break
            print(line.strip())

except FileNotFoundError:
    print("file not found")
except PermissionError:
    print("Not allowed")
finally:
    print("Operation Completed")
    
```

## Task 5

```python

cart = []

while True:
    user_input = input("Enter item price (or 'q' to quit): ")

    if user_input.lower() == 'q':
        break

    try:
        price = float(user_input)   # Convert to float

        if price < 0:
            raise ValueError("Price cannot be negative")

        cart.append(price)
        print(f"Added: {price}")

    except ValueError as ve:
        print("Invalid input:", ve)

# Final summary
print("\n--- Shopping Summary ---")
print("Total items:", len(cart))
print("Total bill:", sum(cart))
```
