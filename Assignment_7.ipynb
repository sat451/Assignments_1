## Task 2

```python
class Product:  #class 
    def __init__(self,name,price,category):
        self.name=name
        self.__price=price
        self.category=category
    
    # getter method
    def get_price(self):
        print(f"The price of product is {self.__price}")
    
    # setter method
    def set_price(self,new_price):
        if int(new_price)>0:
            self.__price=new_price

    def get_info(self):  # method to display the details 
        print(f"{self.name} --> {self.category}")

# objects   

p1=Product("laptop","40000","Electronic")
p2=Product("Mobile","10000","Electronic")
p1.get_info()
p1.get_price()
p2.get_info()
p2.get_price()

# using setter method

p1.set_price("35000")
p1.get_price()
```

## Task 3

```python
class ElectronicProduct(Product):
    def __init__(self,name,price,category,warranty_years):
        super().__init__(name,price,category)
        self.warranty_years=warranty_years

    def get_info(self):
        print(f"{self.name} --> {self.category} --> {self.warranty_years}")

    # creating object of the child class inheriting the properties of Product
    ep1=ElectronicProduct("Monitor","25000","Electronic","5")

    # calling the get_info of the child class showing the method over riding 
    ep1.get_info()
```

## Task 4

```python
class Laptop(Product):

    def get_info(self):   # overriding method
        print(f"Laptop Name : {self.name}")
        print(f"Category : {self.category}")


class Mobile(Product):

    def get_info(self):   # overriding method
        print(f"Mobile Name : {self.name}")
        print(f"Category : {self.category}")
# Creating objects
l1 = Laptop("HP Pavilion","55000","Electronics")
m1 = Mobile("Samsung Galaxy","20000","Electronics")


# List of objects
products = [l1, m1]


# Loop to showiing polymorphism
for item in products:
    item.get_info()
    item.get_price()
    print()
```

## Task 5

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    ## created abstract method 
    @abstractmethod  
    def process_payment(self):
        pass

# subclass of Payment
class CreditCardPayment(Payment): 
    def process_payment(self):
        print("The amount paid by credit card")

# subclass of Payment
class UPIPayment(Payment):
    def process_payment(self):
        print("Tehe amount paid by upi")


p1=CreditCardPayment()
p2=UPIPayment()
p1.process_payment()
p2.process_payment()
```

## Task 6

```python
class Product:  #class 
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category
    
    
    def __str__(self):
        return f"Product({self.name},{self.price},{self.category})"
    def __add__(self,other):
        return self.price + other.price
    


# objects   
product1=Product("laptop",40000,"Electronic")
product2=Product("Mobile",10000,"Electronic")

#for the __str__ method 
print(product1)
print(product2)

# for the __add__method 
conclusion= product1 + product2
print(conclusion)
```

```python
# Inventory Class
class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for p in self.products:
            if p.name == name:
                self.products.remove(p)

    def get_total_value(self):
        total = 0
        for p in self.products:
            total += p.price
        return total

    def show_all_products(self):
        for p in self.products:
            print(p)


# Store Class
class Store:
    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self):
        name = input("Enter product name: ")
        price = float(input("Enter product price: "))
        category = input("Enter category: ")

        product = Product(name, price, category)
        self.inventory.add_product(product)

    def show_summary(self):
        print("\nStore:", self.store_name)
        print("Total Items:", len(self.inventory.products))
        print("Total Value:", self.inventory.get_total_value())


# ---------------- TESTING ----------------

# Create store
store = Store("Tech Store")

# Add 3 products
store.add_new_product()
store.add_new_product()
store.add_new_product()

# Show all products
print("\nAll Products:")
store.inventory.show_all_products()

# Show summary
store.show_summary()

# Using __add__ to combine price
p1 = store.inventory.products[0]
p2 = store.inventory.products[1]

print("\nCombined price of two products:", p1 + p2)
```
