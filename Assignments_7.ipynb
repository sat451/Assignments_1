```python
class Product:
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category
    def get_info(self):
        print(f"The name :{self.name}")
        print(f"The price :{self.price}")
        print(f"The category:{self.category}")
    

car=Product("suzuki","100000","cars")
Mobile=Product("MI","12500","phone")
car.get_info()
Mobile.get_info()
```
